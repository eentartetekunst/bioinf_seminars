# Problem 3

```python
import pandas as pd
```

```python
data_brca = pd.read_csv('BRCA_pam50.tsv', sep='\t', index_col=0)
data_brca.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
svc = SVC(kernel='linear')

data_brca_transformed = data_brca.loc[(data_brca['Subtype'] == 'Luminal A') | (data_brca['Subtype'] == 'Luminal B')]
data_brca_transformed['Subtype'] = le.fit_transform(data_brca_transformed['Subtype'])
data_brca_transformed
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_brca_transformed.loc[:, data_brca_transformed.columns.values[:-1]], data_brca_transformed['Subtype'])
```

```python
svc.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score

preds = svc.predict(X_test)
conf_matrix = confusion_matrix(y_test, preds)
tpr = conf_matrix[0][0] / (conf_matrix[0][0] + conf_matrix[0][1])
tnr = conf_matrix[1][1] / (conf_matrix[1][1] + conf_matrix[0][1])
roc_auc = roc_auc_score(y_test, preds)

print(f"TPR: {tpr}, TNR: {tnr}, ROC-AUC score: {roc_auc}")
print(conf_matrix)
```

```python
most_prominent_features = pd.Series(svc.coef_[0]).abs().sort_values(ascending=False)[:5].index
most_prominent_features
```

```python
data_brca_transformed_trimmed = data_brca_transformed.iloc[:, most_prominent_features]
data_brca_transformed_trimmed['Subtype'] = data_brca_transformed['Subtype']
data_brca_transformed_trimmed.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_brca_transformed_trimmed.loc[:,data_brca_transformed_trimmed.columns.values != 'Subtype'], data_brca_transformed_trimmed['Subtype'])
```

```python
svc.fit(X_train, y_train)
```

```python
preds = svc.predict(X_test)
conf_matrix = confusion_matrix(y_test, preds)
tpr = conf_matrix[0][0] / (conf_matrix[0][0] + conf_matrix[0][1])
tnr = conf_matrix[1][1] / (conf_matrix[1][1] + conf_matrix[0][1])
roc_auc = roc_auc_score(y_test, preds)

print(f"TPR: {tpr}, TNR: {tnr}, ROC-AUC score: {roc_auc}")
print(conf_matrix)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
compressed_brca_data = pca.fit_transform(data_brca_transformed.loc[:, data_brca_transformed.columns.values != 'Subtype'])

X_train, X_test, y_train, y_test = train_test_split(compressed_brca_data, data_brca_transformed['Subtype'])
svc.fit(X_train, y_train)
```

```python
from scipy.stats import norm

norm0 = norm(loc=0)
norm10 = norm(loc=10)

dummy_data = pd.DataFrame({'x': norm0.rvs(size=10000), 'y': norm0.rvs(size=10000), 'class': [0]*10000})
dummy_data = pd.concat([dummy_data, pd.DataFrame({'x': norm10.rvs(size=10000), 'y': norm10.rvs(size=10000), 'class': [1]*10000})])
```

```python
X_train, X_test, y_train, y_test = train_test_split(dummy_data[['x', 'y']], dummy_data['class'])
```

```python
import time

begin = time.time()
svc.fit(X_train, y_train)
end = time.time()
print(end - begin)
```

```python
preds = svc.predict(X_test)
conf_matrix = confusion_matrix(y_test, preds)
tpr = conf_matrix[0][0] / (conf_matrix[0][0] + conf_matrix[0][1])
tnr = conf_matrix[1][1] / (conf_matrix[1][1] + conf_matrix[0][1])
roc_auc = roc_auc_score(y_test, preds)

print(f"TPR: {tpr}, TNR: {tnr}, ROC-AUC score: {roc_auc}")
print(conf_matrix)
```
